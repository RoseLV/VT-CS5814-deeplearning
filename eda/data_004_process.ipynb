```python
import os
import shutil
from PIL import Image
import numpy as np
```

```python
import os
import shutil

# Specify the source folder with files
source_folder = '/home/ran/data/deeplearning/nnUNet_raw/Dataset004/labelsTr'

# Specify the destination folder to store filtered files
destination_folder = '/home/ran/data/deeplearning/nnUNet_raw/Dataset004/output'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Array to store file names with more than 1 distinct value
selected_files = []

# Function to get distinct values in a .tif file
def get_distinct_values(file_path):
    with Image.open(file_path) as img:
        img_array = np.array(img)  # Convert image to a NumPy array
    return np.unique(img_array)

# Loop through the files in the source folder
for file_name in os.listdir(source_folder):
    file_path = os.path.join(source_folder, file_name)
    
    # Check if the item is a .tif file
    if os.path.isfile(file_path) and file_name.lower().endswith('.tif'):
        try:
            distinct_values = get_distinct_values(file_path)
            
            # Check if the file has more than 1 distinct value
            if len(distinct_values) > 1:
                selected_files.append(file_name)
                
                # Copy the file to the destination folder
                shutil.copy(file_path, os.path.join(destination_folder, file_name))
        except Exception as e:
            print(f"Error processing file {file_name}: {e}")

# Print the selected file names
print("Files with more than 1 distinct value:", selected_files)

```

```python
import os
import shutil

# Folder A containing the files
folder_a = '/home/ran/data/deeplearning/nnUNet_raw/Dataset004/imagesTr'

# Folder B where selected files will be copied
folder_b = '/home/ran/data/deeplearning/nnUNet_raw/Dataset004/imagesTr_o'

# Extract prefixes (remove `.tif` from selected_files)
selected_prefixes = [os.path.splitext(file)[0] for file in selected_files]

# Create folder B if it does not exist
os.makedirs(folder_b, exist_ok=True)

# Loop through all files in folder A
for file_name in os.listdir(folder_a):
    file_path = os.path.join(folder_a, file_name)
    
    # Check if the file name starts with any prefix in selected_prefixes
    if any(file_name.startswith(prefix) for prefix in selected_prefixes):
        # Copy the file to folder B
        shutil.copy(file_path, os.path.join(folder_b, file_name))

print(f"Files matching prefixes {selected_prefixes} have been copied to {folder_b}.")
```

```python
import json

# Array of selected files
# selected_files = ['TCGA_FG_A4MT_20020212_12.tif', 'TCGA_HT_8113_19930809_16.tif']

# Generate the desired JSON structure
output_data = [
    {
        "image": f"./imagesTs/{file}",
        "label": f"./labelsTs/{file}"
    }
    for file in selected_files
]

# Specify the output JSON file path
output_file = 'output.json'

# Write the data to a JSON file
with open(output_file, 'w') as json_file:
    json.dump(output_data, json_file, indent=4)

print(f"JSON data has been written to {output_file}")
```
